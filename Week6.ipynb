# 2. Weekly graph question

Suppose you wanted to show the nitrate and phosphate level in a water sample.  What are the advantages and disadvantages of showing this as an area plot, as opposed to two separate line graphs?  How would you adjust the graphs shown to improve the presentation?

**Area Plot**

Advantages: Area plot shows proportional comparison between nitrate and phospate. We can visualize relative contribution over time. This plot effectively highlights the combined level of nitrate and phosphate which is helpful to access eutrophication risk.

Disadvantages: The area plot is not labeled properly and it is harder to compare indivudual trend as series 2 is plotted on top of series 1. Area plot mainly displays additive behavior and it can mislead the interpretation by overlooking individual contribution of the chemical components. Also exact values are harder to interpret.

**Two separate line graphs:**

Advantages: Charts are porperly labeled and individual trends for nitreate and phosphate are clearly visible. There are no stacking distortion. These plots are better for overall threshold comparison.

Disadvantages: Primary disadvantge will be we have to compare two seperate charts to access the chemical components. Harder to visualize cumulative nutrient load(nitrate and phosphate combined).

**Improvemnts:**

Area graph: Add clear labels and title to communicate the data. Use the transparency via alpha parameter.

Line graphs: Combine phosphate and nitrate trend lines into one line plot. Add a horizontal eutrophic threshold line. Add legends and a title to make it more presentable to the audience.

```python
datasource= "http://archive.ics.uci.edu/static/public/17/data.csv"

df_BreastCancer = pd.read_csv(datasource, sep=',', header=0) #loading breast cancer dataset
df_BreastCancer.sample(10, random_state=42)
```

```python
#Find correlations between pairs of variables:
#1. Compute the correlation matrix
df_BreastCancer_2=df_BreastCancer.drop(columns=["ID","Diagnosis"])
Corr=df_BreastCancer_2.corr()
#print(Corr)

#2.Find top correlation between pairs
corr_flatten=Corr.unstack()
#remove self correlation
corr_flatten=corr_flatten[corr_flatten.index.get_level_values(0)!=corr_flatten.index.get_level_values(1)]
#corr_top_pairs=corr_flatten.abs().sort_values(ascending=False).head(70)
corr_top_pairs=corr_flatten.abs().sort_values(ascending=False).head(60)
print(corr_top_pairs)
```

```python
from numpy._core.defchararray import title
import matplotlib.pyplot as plt
#Draw scatterplots, especially when the correlation is large
#1. Compute the correlation matrix
df_BreastCancer_3=df_BreastCancer.drop(columns=["ID"]) #drop ID
df_BreastCancer_3["Diagnosis"]=df_BreastCancer_3["Diagnosis"].map({"M":1,"B":0}) #convert dignosis to numeric
Corr=df_BreastCancer_3.corr()

#2.Find top correlation between pairs
corr_flatten=Corr.unstack()
#remove self correlation
corr_flatten=corr_flatten[corr_flatten.index.get_level_values(0)!=corr_flatten.index.get_level_values(1)]
#sort the correlation
corr_sorted_pairs=corr_flatten.abs().sort_values(ascending=False)
#deduplicated data
corr_sorted_pairs=corr_sorted_pairs[~corr_sorted_pairs.index.duplicated()]
#large correlated pairs excluding diagnosis:
corr_top= [x for x in corr_sorted_pairs.index if "Diagnosis" not in x ][:10]
print(corr_top)

#Genrating scatter plots
for p1, p2 in corr_top:
  plt.figure(figsize=(5,5))
  sns.scatterplot(x=df_BreastCancer_3[p1], y=df_BreastCancer_3[p2], hue=df_BreastCancer_3["Diagnosis"], palette=["blue","green"],alpha=0.6)
  plt.title(f"{p1} VS {p2}")
  plt.xlabel(p1)
  plt.ylabel(p2)
  plt.legend(title="(0=Benign,1=Malignant)")
  plt.tight_layout()
  plt.show()

```

**Can you think of any confounding variables?**
This dataset contains several confounding variables.

1.Malignacy Diagnosis: The target variable diagnosis is one of the major confounder. Malignant tumor cells tend to be larger in size with more irregular shape than benign cells. These cells also have more concave points and higher concavity compared to benign cells. So diagnosis can explain positive correlations in (area, compactness), (radius, concave_points), (radius, concavity).

2.Tumor size: This is also another major confounder. Tumor growth leads to corresponding increases in its radius, perimeter and area and it also increases the shape irregularity. Larger tumors tend to generate higher concavity, more concave points and greater compactness. So it explains correlation between radius, perimeter, area, concavity.
