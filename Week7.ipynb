# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

Answer:
Possibilities 1: Use different colors to highlight last three months.

Possibilities 2: Increase marker size for last three months.

Possibilities 3: Add shading for last three months.

Possibilities 4: Add vertical lines to create a visual boundary for last three
                 months.

Possibilities 5: Add labels or annotations to point viewers attention.

Possibilities 6: Change line thickness for the recent period.


```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)


ax = sns.lineplot(x = datearray, y = time_series, color="skyblue", marker='o')

#Possibilites 1 Use different colors to highlight last three months.
ax.plot(datearray[-3:],time_series[-3:], color="green", linewidth=3, marker='o')
#Possibilities 2: Increase marker size for last three months.
ax.scatter(datearray[-3:],time_series[-3:], color="green", linewidth=3, s=20)

ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')

ax = sns.lineplot(x = datearray, y = time_series, color="gray", marker='o')

#Possibilites 1 Use different colors to highlight last three months.
ax.plot(datearray[-3:],time_series[-3:], color="green", linewidth=3, marker='o')
#Possibilities 2: Increase marker size for last three months.
ax.scatter(datearray[-3:],time_series[-3:], color="green", linewidth=3, s=20)

ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```
