# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

Answer:
Possibilities 1: Use different colors to highlight last three months.

Possibilities 2: Increase marker size for last three months.

Possibilities 3: Add shading for last three months.

Possibilities 4: Add vertical lines to create a visual boundary for last three
                 months.

Possibilities 5: Add labels or annotations to point viewers attention.

Possibilities 6: Change line thickness for the recent period.


```python
#updating chart1(Nitrate level)
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)


ax = sns.lineplot(x = datearray, y = time_series, color="skyblue", marker='o')

#Use different colors to highlight last three months.
#Change line thickness for last three month
ax.plot(datearray[-3:],time_series[-3:], color="green", linewidth=3, marker='o')
#Increase marker size for last three months.
ax.scatter(datearray[-3:],time_series[-3:], color="green", linewidth=3, s=20)
#Add shading for last three months.
ax.axvspan(datearray[-3],datearray[-1], color="gray", alpha=0.3)
#Add vertical lines to create a visual boundary for last three months.
ax.axvline(datearray[-3], linestyle="--", color="black")
#Add labels or annotations to point viewers attention.
ax.text(datearray[-2],104,"Lastest three", ha="center")

ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
#updating chart2(Phosphate level)
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')

ax = sns.lineplot(x = datearray, y = time_series, color="gray", marker='o')

#Possibilites 1 Use different colors to highlight last three months.
ax.plot(datearray[-3:],time_series[-3:], color="green", linewidth=3, marker='o')
#Possibilities 2: Increase marker size for last three months.
ax.scatter(datearray[-3:],time_series[-3:], color="green", linewidth=3, s=20)
#Possibilities 3: Add shading for last three months.
ax.axvspan(datearray[-3],datearray[-1], color="skyblue", alpha=0.3)
#Possibilities 4: Add vertical lines to create a visual boundary for last three months.
ax.axvline(datearray[-3], linestyle="--", color="black")
#Add labels or annotations to point viewers attention.
ax.text(datearray[-2],107,"Lastest three", ha="center")

ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
#Figure 4.4 Preventive attributes
from matplotlib.ticker import PercentFormatter
import matplotlib.pyplot as plt
#Generating Data for figure 4.4 preventive attributes(Storytelling with data)
df_concerns = pd.DataFrame(columns = ["concerns"])

df_concerns["concerns"] = ["Engine power is less than expected", "Tires make excessive noise while driving", "Engine makes abnormal/excessive noise",
                                             "Seat materials concerns", "Excessive wind noise","Hesitation or delay when shifting", "Bluetooth system has poor sound quality",
          "Steering system/wheel has too much play", "Bluetooth system is difficult to use","Front seat audio/entertainment/navigation controls" ]
df_concerns["estimate"] = [12.9,12.3,11.6,11.6, 11.0,10.3,10.0,8.8,8.6,8.2]

#sort by estimate
df_concerns.sort_values("estimate")

# Custom color to highlight top 10 with red as shown in the book
color=["#800000" if i>=10 else "gray" for i in df_concerns["estimate"]]

sns.set_theme(style="white")
ax=sns.barplot(data=df_concerns, x="estimate", y="concerns",palette=color)

#removing x ticks and positioning the xlabel to the top as shown in the book
ax.xaxis.set_label_position("top")
ax.xaxis.set_ticks([])

#to remove the spines as shown in the book
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["top"].set_visible(False)


#Add values on the plot as shown in the book
for i, j in enumerate(df_concerns["estimate"]):
  ax.text(j-1.3, i, f"{j}", va="center", color="white")

#Add texts
ax.text(-9,-1,"Top 10 design concerns", color="black")
plt.xlabel("Concerns per 1,000", loc="left")
ax.text(-9,-2,"\n7 of the top 10 design concerns have 10 or more concern per 1,000.", color="#800000")
plt.ylabel(" " )
plt.show()
```
