# 2. Weekly graph question

Below are a histogram and boxplot representation of the same data. A pharmacy is keeping a record of the prices of the drugs that it sells, and an administrator wants to know how much the more expensive drugs tend to cost, in the context of the other prices.

Please write a short explanation of the pros and cons of these two representations. Which would you choose? How would you modify the formatting, if at all, to make it more visually interesting, clear, or informative?

```python
import numpy as np
import pandas as pd

np.random.seed(0)
num_data = 100
data = np.exp(np.random.uniform(size = num_data) * 4)
df = pd.DataFrame(data.T, columns = ["data"])
df
```

```python
print("The 75th percentile is:", df.quantile(q = 0.75))
df.plot.hist()
```

```python
df.plot.box()
```

**Histogram with 75 percentile data:**

   Pros: Shape of the distribution is clearly visible. Cluster of drug price are shown. Outliers(very low and very high drug price) are visible. Since print statement is added 75 percentile data is reported clearly.
   Cons: Exact data summary(median, quartile) is not visible. Choice of bin can change the appearence. 75% is not visible on the chart without additional note or print statement.

**Box Plot:**
   Pros: It shows median, quartile and outliers clearly. 75th percentile is visible on top of the box. 25th percentile is visible at bottom of the box, line inside the box shows the median. It is easier to find out most expensive drugs.
   Cons: Distribution shape is not available and it will not show cluster of drugs price.

 **Which would I choose?**
 For drug cost analysis I would like to choose the 2nd visual-Boxplot.

**Modify the formatting**:
  **Histogram**: By adding- a proper title(Drug price distribution), clearly label x and y axis (drug price, frequency), a correct bin type and a line for 75 percentile and median.
  **Box Plot:** By adding- a proper title(Drug price distribution) and x axis label(Drug Price) and probably add a showmean marker.

```python
#Modify Histogram
import matplotlib.pyplot as plt
p75=df["data"].quantile(0.75)
median= df["data"].median()
print("The 75th percentile is:", df.quantile(q = 0.75))
plt.hist(df["data"], bins=15, alpha=0.6, color="blue",label="Data")

#plot 75% percentile
plt.axvline(p75,color="gray", linestyle="--",linewidth=2,label=f'75th percentile ({p75:.2f})')
#plot median
plt.axvline(median,color="black", linestyle="--",linewidth=2,label=f'median ({median:.2f})')
plt.title("Drug price distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
#Modify the Boxplot:
p75=df["data"].quantile(0.75)
median= df["data"].median()
plt.figure(figsize=(6,5))
plt.boxplot(df["data"], showmeans=True,meanprops=dict(marker='o',markerfacecolor="red",markersize=8),
            medianprops=dict(color="green", linewidth=2))
#adding textbox to show median and 75th percentile
text=f"Median: {median:.2f}\n75th percentile: {p75:.2f}"
plt.text(1.15,45,text,fontsize=10,bbox=dict(facecolor="white",alpha=0.6,edgecolor="black"))
plt.xlabel("Price")
plt.title("Drug price distribution")
plt.show()
```

```python
#Draw histograms for each feature or column
import matplotlib.pyplot as plt

#selecting numeric columns for histogram
numeric_only=df_BreastCancer_2.select_dtypes(include="number")

for c in numeric_only.columns:
  plt.figure(figsize=(5,5))
  plt.hist(numeric_only[c],bins=20)
  min_val=numeric_only[c].min()
  max_val=numeric_only[c].max()
  #show min
  plt.scatter(numeric_only[c].min(),0,color="red",s=100)
  plt.text(min_val,4,f"min {min_val:.4f}",color="black")
   #show max
  plt.scatter(numeric_only[c].max(),0,color="yellow",s=100)
  plt.text(max_val,4,f"max {max_val:.4f}",color="black")

  plt.title(f"Histogram of {c}")
  plt.tight_layout
  plt.legend()
  plt.show()
```

```python
#Draw histogram variants for each feature or column
#Swarm plot
import seaborn as sns
numeric_only_cols=df_BreastCancer_2.select_dtypes(include="number").columns

for c in numeric_only_cols:
  plt.figure(figsize=(4,4))
  sns.swarmplot(y=df_BreastCancer_2[c])
  plt.title(f"Swarm plot of {c}")
  plt.tight_layout()
  plt.show()
```

```python
#draw kde plots for all columns
numeric_only_cols=df_BreastCancer_2.select_dtypes(include="number").columns

for c in numeric_only_cols:
  plt.figure(figsize=(4,4))
  sns.kdeplot(data=df_BreastCancer_2, x=c)
  plt.title(f"kde plot of {c}")
  plt.tight_layout()
  plt.show()
```

```python
#draw violin plots for all columns
numeric_only_cols=df_BreastCancer_2.select_dtypes(include="number").columns

for c in numeric_only_cols:
  plt.figure(figsize=(4,4))
  sns.violinplot(y=df_BreastCancer_2[c])
  plt.title(f"Violin plot of {c}")
  plt.tight_layout()
  plt.show()
```

```python
#Draw grouped histogram
selected_cols= ["radius1","area1","concavity1","fractal_dimension1", "compactness1","area2","concave_points3","area3"]
for c in selected_cols:
  plt.figure(figsize=(6,4))
  sns.histplot(data=df_BreastCancer_2, x=c, hue="Diagnosis", bins=20,stat="density")
  plt.title(f"Histogram of {c}")
  plt.tight_layout()
  plt.show()
```

# 4. Storytelling With Data graph

Reproduce any graph of your choice in p. 52-68 of the Storytelling With Data book as best you can.  (The second half of chapter two).  You do not have to get the exact data values right, just the overall look and feel.

```python
#Fig2.27 option 2 pull apart vertically
#Generating Sales Revenue data
from matplotlib.ticker import StrMethodFormatter
num_employee = 50
np.random.seed(0)

df_sale = pd.DataFrame(columns = ["employeeID"])
df_sale["employeeID"] = [str(x) for x in np.arange(num_employee)]
df_sale["year"] = np.random.choice(["2013","2014"], size = num_employee)
df_sale["Quarter"] = np.random.choice(["Q1", "Q2", "Q3", "Q4"], size = num_employee)
df_sale["Sales"] = np.random.randint(20,100,50)
df_sale["price"] = np.random.uniform(2,100,50).round(2)
df_sale["revenue"] = (df_sale["price"] * df_sale["Sales"]).round(2)
df_sale

#group the data by yr_qtr:
df_sale["Yr_Qtr"]=df_sale["year"]+" "+df_sale["Quarter"]
agg=df_sale.groupby("Yr_Qtr").agg({"revenue": "sum", "Sales":"sum"}).reset_index()
#agg

fig, (ax1,ax2)=plt.subplots(2,1, figsize=(6,7),sharex=True)
fig.subplots_adjust(hspace=0)

#Top chart trend line for sales:
ax1.plot(agg["Yr_Qtr"],agg["Sales"],marker="o", color="blue", label="Sales")
ax1.set_ylabel("#of sales employee")
ax1.tick_params(bottom=False)
ax1.grid(False)
ax1.spines["bottom"].set_visible(False)
ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)
ax1.set_title(" Sales and Revenue by Year Qtr")

#Bottom chart bar graph for revenue:
ax2.bar(agg["Yr_Qtr"],agg["revenue"], color="skyblue", label="Revenue")
ax2.set_ylabel("Revenue($)")
ax2.set_xlabel("Year Qtr")
ax2.yaxis.set_major_formatter(StrMethodFormatter('${x:,}'))
ax2.spines["top"].set_visible(False)
ax2.spines["right"].set_visible(False)
ax2.grid(False)

plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()



```
