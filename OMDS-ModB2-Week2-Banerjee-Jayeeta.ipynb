```python
#Check outliers using quantile method
for column in ["radius1","texture1", "concavity1", "area1"]:
  Q1= df_BreastCancer[column].quantile(0.25)
  Q2= df_BreastCancer[column].quantile(0.75)
  IQR=Q2-Q1

  lower=Q1-2*IQR
  upper=Q2+2*IQR

  outliers= df_BreastCancer[(df_BreastCancer[column]<lower) | (df_BreastCancer[column]>upper)]
  print("priting outliers:")
  print(outliers[[column]])
```

```python
#Generating Data for Figure 0.5 Example 2 (Storytelling with data)
import matplotlib.pyplot as plt
num_students = 100
np.random.seed(0)
df_survey = pd.DataFrame(columns = ["studentID"])
df_survey["studentID"] = [str(x) for x in np.arange(num_students)]
df_survey["Program"] = np.random.choice(["Before","After"], size = num_students)
df_survey["survey_data"] = np.random.choice(["Bored", "Not Great", "OK", "Kind of interested", " Excited"], size = num_students)
#print(df_survey)

#students feedback grouped by program and survey
survey_counts= df_survey.groupby(["survey_data","Program"]).size().unstack(fill_value=0)
#calculate percentage
survey_percentage=survey_counts.div(survey_counts.sum(axis=0), axis=1)*100
#plot the bar graphs
bar_chart=survey_percentage.plot(kind='bar', figsize=(8,6))
#Adding percentage label
for c in bar_chart.containers:
  bar_chart.bar_label(c, fmt='%.1f%%', label_type="edge")
plt.xlabel("\nBased on survey of 100 students conducted before and after pilot program", loc="left")
plt.title("How do you feel about science?", loc="left")
plt.xticks(rotation=0)

plt.legend(title="program")

ax=plt.gca()
#add before after notes within chart:
ax.text(0.2, 36,"BEFORE program many children \nfelt BORED about science", ha="center",va="center", color="black")
ax.text(2.7, 30,"AFTER program more children \nfelt Kind of Interested and \nExcited about science", ha="center",va="center", color="blue")
#to remove the spines as shown in the book
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.set_yticks([])

plt.tight_layout()
plt.show()
```
