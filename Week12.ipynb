# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

**Answer:**

Plot 1(two side by side sub plots): **Advantages:** It reduces visual clutter by comparing the both trendlines in one plot side by side with same horizontal scale(aligned x axis). It avoids overlapping which can be difficult to interpret. It is helpful for independent pattern recognition within each week, such as mid week peak or drop in weekend can be easily seen. **Disadvantages:** Potential inconsistency in y-axis scaling can be noticed. As the y-axis limits are not shared the differences can be misleading. Subplots are not correctly labeled, it is difficult to understand which subplots corresponds to which week. It is harder to directly compare the two datasets beetween weeks as lines are not on the same axis. Additionally axises are not poperly labeled.

plot 2(both weeks on the same axis): **Advantages** In this plot the two trendlines are sharing both x-axis and y-axis scales, so we can easily compare the data for the same day within the week without any risk of misleading differences due to different axis limits. Plots are more compact and axises are properly labeled. **Disadvantages:** Legends are missing, it is unclear which line correspons to which week. Independent trends can be harder to analyze.


```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
#Chapter 8 fig: 8.4 Change to line graph

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

#dataset
df_sale=pd.DataFrame(columns = ["Year"])
df_sale["Year"] = [2019,2020,2021,2022,2023,2024,2025]
df_sale["Product A"] = [400,408,410, 395, 360, 290, 270 ]
df_sale["Product B"] = [350,358,362, 350, 280, 260, 240 ]
df_sale["Product C"] = [0,0,150, 170, 200, 250, 170 ]
df_sale["Product D"] = [0,0,0, 160, 270, 220, 211 ]
df_sale["Product E"] = [0,0,0, 0, 0, 155, 230 ]

#replacing 0 values with nan for clear visualization
df_sale=df_sale.replace(0,np.nan)
#df_sale

plt.figure(figsize=(11,6))
products= df_sale.columns[1:]
x_positions = np.arange(len(products))

for i, p in enumerate(products):
  y= df_sale[p].values
  x=np.full_like(y, x_positions[i], dtype=float)
  x=x + np.linspace(-0.2,0.2,len(y))
  plt.plot(x,y, linewidth=2, color="gray")
  idx=np.where(~np.isnan(y))[0]
  if len(idx) > 0:
    start = idx[0]
    end = idx[-1]
    plt.text(x[start], y[start]-8, str(df_sale["Year"][start]),
             fontsize=9, ha="center", alpha=0.7)

    plt.text(x[end], y[end]-8, str(df_sale["Year"][end]),
             fontsize=9, ha="center", alpha=0.7)

plt.xticks(x_positions,products)
plt.ylim(bottom=0)
plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))
plt.title("Average Retail Product Price Per Year")
plt.grid(axis='y', linestyle='--',alpha=0.6)

plt.tight_layout()
plt.show()
```
